1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns = ["Unnamed: 0"], errors="ignore")
df.info()
```

```python
#Standardize column names

df.columns = df.columns.str.lower().str.replace(" ","_")
df.info()
```

```python
#Clean empty rows

df = df.dropna(how="all")
df.info()
```

```python
df['response'].unique()
```

```python
'''
New DataFrame that only includes customers who:
have a low total_claim_amount (e.g., below $1,000),
have a response "Yes" to the last marketing campaign.
'''

filtered_df = df [
    (df["total_claim_amount"] < 1000)&
    (df["response"]=="Yes")
    ]

filtered_df.head()
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
#Customer that responded "Yes"
yes_df = df[df["response"]=="Yes"]
```

```python
#Average monthly_premium vs customer lifetime value vs total_claim_amount
premium_summary= (
    yes_df.groupby(["policy_type","gender"])
    .agg({
        "monthly_premium_auto":"mean",
        "customer_lifetime_value":"mean",
        "total_claim_amount":"mean"
    })
    
    .round(2)
    .reset_index()
)
premium_summary
```

```python
'''The most profitable segment -> Female Personal Auto with highest monthly premium and highest clv.
Second best -> Male Special Auto with high clv and second lower claim amount.

Males with Corporate Auto are more stable with less claim ampunts of all.

'''
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
'''No. customers - policies - each state (>500 customers)'''
state_customers = df.groupby("state")["customer"].count().reset_index()
state_customers = state_customers .rename(columns={"customer":"total_customers"})
state_customers
```

```python
#filter more than 500
state_customers_500 = state_customers[state_customers["total_customers"]>500]
state_customers_500
```

```python
#Sort
state_customers_500 = state_customers_500.sort_values(
    by = "total_customers",
    ascending = False
)
state_customers_500
```

```python
'''
California has the most customers, having the highest amount of customer distribution there.

Washington and Nevada can be an opportunity to increase the market.
'''
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
clv_statistics = (
    df.groupby(["education","gender"])["customer_lifetime_value"]
    .agg(["max","min","median"])
    .round(2)
    .reset_index()
)

clv_statistics
```

```python
'''
Educational level seems to have more impact on customer lifetime value than gender.

There are significant outliers in some segments(ex. Bachelor F, High School or Below M), 
indicating the presence of a few very high-value customers.

Customers with 'High School or Below' and 'College' education levels 
have the highest median CLV, indicating they are more valuable on average.

Customers with a Doctor degree show lower median values.

'''
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
#Policies (customer) - state/month - Top 3
state_month = (
    df.groupby(["state","month"])["customer"]
    .count()
    .reset_index(name="num_policies")
)

state_month
```

```python
policies_sold = (
    state_month.groupby("state")["num_policies"]
    .sum()
    .reset_index()
)

policies_sold
```

```python
policies_top = policies_sold.sort_values(
    by="num_policies",
    ascending=False
).head(3)
policies_top
```

```python
top_policies_pivot = (
    state_month
    .pivot(index= "state", columns="month",values="num_policies")
    .loc[policies_top["state"]]
)
top_policies_pivot 
```

7. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by marketing channel.

External Resources for Data Filtering: https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9

```python
#marketing channels/customer_response_rate
df["sales_channel"].unique()
```

```python
df_channels = pd.get_dummies(df,columns=["sales_channel"]) 
df_channels 
```

```python
melted = df_channels.melt(
    id_vars = ["response"],
    #Seleccionar las que tienen sales_channel que son las que quiero convertir a filas
    value_vars = [col for col in df_channels.columns if "sales_channel" in col],
    var_name="channel", #nombre a la columna de los nuevos valores
    value_name="is_channel" #nombre a la columna de los nuevos valores de su lista
)

melted #Convierte algo ancho (columnas) en algo largo (filas)
```

```python
melted = melted[melted["is_channel"]== 1]
melted #Filtrar solo las que tienen el canal (True o 1)
```

```python
response_rate = (
    melted.groupby("channel")["response"]
    .apply(lambda x:(x=="Yes").mean())
    .reset_index(name="response_rate")
)

response_rate #% of customers that answered Yes
```
